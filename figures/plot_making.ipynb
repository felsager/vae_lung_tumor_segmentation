```python
import numpy as np
import torch
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vae_train_loss = torch.load('../losses/vae_train_losses.pt')
vae_dev_loss = torch.load('../losses/vae_dev_losses.pt')

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(vae_train_loss, label='Train Loss', linewidth=2, color='#1974D2')
ax.plot(vae_dev_loss, label='Dev Loss', linewidth=2, color='#FF007F')
ax.legend(['Train Loss', 'Dev Loss'])
ax.set_title('VAE Losses')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
plt.tight_layout()
plt.savefig('vae_losses.svg', format='svg', dpi=1200)
```

```python
seg_train_loss = torch.load('../losses/seg_train_losses.pt')
seg_dev_loss = torch.load('../losses/seg_dev_losses.pt')

fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(seg_train_loss, label='Train Loss', linewidth=2, color='#1974D2')
ax.plot(seg_dev_loss, label='Dev Loss', linewidth=2, color='#FF007F')
ax.legend(['Train Loss', 'Dev Loss'])
ax.set_title('Segmentation Decoder Losses')
ax.set_xlabel('Epoch')
ax.set_ylabel('Loss')
plt.tight_layout()
plt.savefig('seg_losses.svg', format='svg', dpi=1200)
```
